```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, HuberRegressor

# Tạo dữ liệu mô phỏng
np.random.seed(0)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10, 30])  # Điểm (6, 30) là outlier

# Hồi quy tuyến tính thường (OLS)
ols = LinearRegression().fit(X, y)
y_pred_ols = ols.predict(X)

# Hồi quy mạnh (Huber)
huber = HuberRegressor().fit(X, y)
y_pred_huber = huber.predict(X)

# Vẽ đồ thị
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='black', label='Dữ liệu gốc')
plt.plot(X, y_pred_ols, color='blue', label='OLS Regression')
plt.plot(X, y_pred_huber, color='red', linestyle='--', label='Huber Regression')
plt.title("So sánh Hồi quy OLS và Hồi quy mạnh (Huber)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
